```python
import numpy as np
def jacobi(A, b, x0=None, tol=1e-2, max_iterations=10)->np.ndarray:
    n = len(b)
    x = np.zeros_like(b) if x0 is None else x0.copy()  
    x_new = x.copy()
    for it_count in range(max_iterations):
        for i in range(n):
            sum1 = np.dot(A[i, :i], x[:i])
            sum2 = np.dot(A[i, i + 1 :], x[i + 1 :])
            x_new[i] = (b[i] - sum1 - sum2) / A[i, i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            break
        x = x_new.copy()
    return x_new

def gauss_seidel(A, b, x0=None, tol=10e-3, max_iterations=10):
    n = len(b)
    x = np.zeros_like(b) if x0 is None else x0.copy()  
    for it_count in range(max_iterations):
        x_old = x.copy()

        for i in range(n):
            sum1 = np.dot(A[i, :i], x[:i])
            sum2 = np.dot(A[i, i + 1 :], x_old[i + 1 :])
            x[i] = (b[i] - sum1 - sum2) / A[i, i]
        if np.linalg.norm(x - x_old, ord=np.inf) < tol:
            break
    return x

a = np.array([[3,-1,1],[3,6,2],[3,3,7]], dtype=float)
b_a = np.array([1, 0, 4], dtype=float)
solution = gauss_seidel(a, b_a, x0=None, tol=10e-3, max_iterations=2)
print("Soluci贸n correcta seidel:", solution)
solution = jacobi(a, b_a, x0=None, tol=10e-3, max_iterations=2)
print("Soluci贸n correcta jacobi:", solution)
```
