# Ejercico 1

Suma infinita

```python

v_real =2
S=0
n=0
while True:
        i = 1/(2**n)
        S += i
        n += 1 
        print(f"termino{n}:{i},suma parcial:{5}")
        error_abs = abs (v_real - S)
        if error_abs > 10**-1:
            break
        print(f"suma:{S},Numero de terminos:{n},Error absoluto:{error_abs}")
```

# EJERCIO 2
Buble sort

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        #
        for j in range(0, n - i - 1):
   
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

numeros = [2,3,5,8,4,1]

print("Arreglo original:", numeros)
bubble_sort(numeros)
print("Arreglo ordenado:", numeros)
```

```python
import matplotlib.pyplot as plt


n = 1000
fibonacci = [0, 1]

for i in range(2, n+1):
    fibonacci.append(fibonacci[-1] + fibonacci[-2])


ratios = [fibonacci[i+1]/fibonacci[i] for i in range(1, len(fibonacci)-1)]


phi = (1 + 5 ** 0.5) / 2

# --- Graficar ---
plt.figure(figsize=(10, 6))
plt.plot(range(2, len(fibonacci)), ratios, marker='o', label='RazÃ³n Fibonacci')
plt.axhline(y=phi, color='r', linestyle='--', label=f'NÃºmero Ã¡ureo Ï â {phi:.5f}')

plt.title('AproximaciÃ³n al nÃºmero Ã¡ureo con la sucesiÃ³n de Fibonacci')
plt.xlabel('n (posiciÃ³n en la sucesiÃ³n)')
plt.ylabel('F(n+1) / F(n)')
plt.legend()
plt.grid(True)
plt.show()
```
